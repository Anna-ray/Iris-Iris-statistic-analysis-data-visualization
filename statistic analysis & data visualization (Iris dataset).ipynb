```python
import pandas as pnd
import matplotlib.pyplot as plt

data = pnd.read_csv(r"iris.csv",header="infer")
data.info()
print(data.head(n=10))
print(data.mean(axis='index'))                                          
print(data.var(axis='index'))                 
print(data.std(axis='index')) 
print(data.min(axis='index')) 
print(data.max(axis='index')) 
x=data["variety"] 
print(x.value_counts(normalize=True))
data.describe()

c1=data.corr()   
c2=data.cov()     
plt.imshow(c1)
plt.imshow(c2)

plt.matshow(data.corr())
plt.show()

plt.hist(data.corr())
plt.show()
```

```python
import seaborn as sns
sns.boxplot(data['sepal.width'],data['variety'])
sns.boxplot(data['sepal.length'],data['variety'])
plt.xlabel('Sepal Width    Sepal Length')
```

```python
sns.boxplot(data['petal.width'],data['variety'])
sns.boxplot(data['petal.length'],data['variety'])
plt.xlabel('Petal Width           Petal Length')
```

```python
sns.scatterplot(data['sepal.length'],data['sepal.width'],data['variety'])
plt.legend(loc='upper left')
plt.title('Spal Length/Width')
```

```python
sns.scatterplot(data['petal.length'],data['petal.width'],data['variety'])
plt.legend(loc='upper left')
plt.title('Petal Length/Width')
```
